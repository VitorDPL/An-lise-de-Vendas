```python
import pandas as pd
```

```python
df = pd.read_excel('relatorio_vendas_robusto.xlsx')

display(df.head())
```

```python
# região com o maior numero de cancelamentos
regiao_com_mais_cancalementos = (df[df['Status do Pedido'] == 'Cancelado']['Região'].value_counts())
print("Região com o maior número de cancelamentos")
print(regiao_com_mais_cancalementos)

print("---------------------------------------------------------------------------")

# região com o maior numero de clientes insatisfeitos
regiao_com_mais_clientes_insatisfeitos = (df[df['Feedback do Cliente'] == 'Muito Insatisfeito']['Região'].value_counts())
print("Região com o maior número de clientes insatisfeitos")
print(regiao_com_mais_clientes_insatisfeitos)

print("---------------------------------------------------------------------------")

# produto com maior numero de insatisfações
produtos_com_mais_insatisfacoes = (df[df['Feedback do Cliente'] == 'Insatisfeito']['Produto'].value_counts())
print("Produto com maior numero de insatisfações")
print(produtos_com_mais_insatisfacoes)

print("---------------------------------------------------------------------------")
```

```python
cancelados = df[df['Status do Pedido'] == 'Cancelado']

bins = [0, 10, 15, 20, 30, 50, 100] 
labels = ['0-10', '10-15', '15-20', '20-30', '30-50', '50-100']
cancelados['Faixa de Frete'] = pd.cut(cancelados['Frete'], bins=bins, labels=labels)

cancelamentos_por_faixa = cancelados['Faixa de Frete'].value_counts().sort_index()

frete_medio_por_faixa = cancelados.groupby('Faixa de Frete')['Frete'].mean()

dados_corr = pd.DataFrame({
    'Frete Médio': frete_medio_por_faixa,
    'Cancelamentos': cancelamentos_por_faixa
}).reset_index()

print("Tabela de faixas de frete e cancelamentos:")
print(dados_corr)

# percebe-se que entre 7 a 17 a taxa de cancelamento é parecida, valores superiores tendem a ser mais cancelados

```

```python
# produto mais vendido
produtos_mais_vendido = df['Produto'].value_counts()
print("Produtos mais vendidos")
print(produtos_mais_vendido)

print("---------------------------------------------------------------------------")

produtos_menos_vendido = df['Produto'].value_counts().sort_values(ascending=True)
print("Produtos menos vendidos")
print(produtos_menos_vendido)
```

```python
# clientes que compraram a calça jeans e que estão insatisfeitos
clientes_que_compraram_calca_jeans = df[df['Produto'] == 'Calça Jeans']
clientes_que_compraram_calca_jeans_e_estao_insatisfeitos = (clientes_que_compraram_calca_jeans[clientes_que_compraram_calca_jeans['Feedback do Cliente'] == 'Insatisfeito']).value_counts()

print("Clientes que compraram a calça jeans e estão insatisfeitos")
print(clientes_que_compraram_calca_jeans_e_estao_insatisfeitos)
```
